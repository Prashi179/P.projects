```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Creating dataframe, using head to see if it worked
df=pd.read_csv("Instagram-Data.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# returning the relative frequency of audience country elements
# can see that 14% of the top 1,000 influencers have the India as their audience country

df["Audience Country"].value_counts(normalize=True).round(2)
```

```python
# creating a pivot table to show statistics for categories
df.pivot_table(index="Category",values="Followers",aggfunc=["mean","min","max","sum"],sort=False)
```

```python
# creating a new dataframe for the India
df_India=df[df["Audience Country"]=="India"]
df_India.head()
```

```python
# turning the floats to ints
df_India = df_India.astype({"Followers":"int", "Authentic engagement":"int", "Engagement avg": "int"})
df_India.head()
```

```python
# pivot table to show statistics on category followers in the India
df_India.pivot_table(index= 'Category', values= "Followers", aggfunc= ['mean', 'sum','min','max'], sort= False)
```

```python
# showing the Category with the most followers
tf=df_India.nlargest(10,"Followers")
tf.pivot_table(index="Category",values="Followers",aggfunc=["sum"], sort=False)
```

```python
# creating dataframe for top 20 followers
t2f=df_India.nlargest(20,"Followers")
t2f.pivot_table(index="Category",values="Followers",aggfunc=["sum"],sort=False)
```

```python
# creating dataframe for top 20 followers
t2f=df_India.nlargest(20,"Followers")
t2f
```

```python
# showing the categories with the most avg audience engagement
avgtopeng=df_India.nlargest(20,"Engagement avg")
avgtopeng.pivot_table(index="Category",values="Engagement avg",sort=False)
```

```python
# creating a new fashion category dataframe to include all of the subsets
fashion=df_India[df_India["Category"].str.contains("Fashion", na=False)]
fashion.head()
```

```python
# plotting followers
plt.figure(figsize=(12,8))
fig1=plt.bar(tf.Category, tf.Followers,)
plt.xticks(fontsize= 12, rotation= 90)
plt.ylabel("Followers", fontsize = 16)
plt.xlabel("Category", fontsize = 16)
plt.title("Top 10 Categories by Followers", fontsize = 15)
plt.tight_layout()
# remove scientific notation
plt.ticklabel_format(axis= "y", style= 'plain')
plt.show()
```

```python
# plotting followers
plt.figure(figsize=(12,8))
fig1=plt.bar(t2f.Category, t2f.Followers,)
plt.xticks(fontsize= 12, rotation= 90)
plt.ylabel("Followers", fontsize = 16)
plt.xlabel("Category", fontsize = 16)
plt.title("Top 20 Categories by Followers", fontsize = 15)
plt.tight_layout()
# remove scientific notation
plt.ticklabel_format(axis= "y", style= 'plain')
plt.show()
```

```python
# plotting the number of accounts between 0 and 60 million followers
plt.figure(figsize=(12,8))
plt.hist(df_India.Followers, range=(0, 60000000))
plt.xlabel('Followers', fontsize = 16)
plt.ylabel('Number of Instagram Accounts', fontsize = 16)
plt.title('Histogram: Follower Counts of Top Instagram Accounts in the India', fontsize = 20)
plt.tight_layout()
# remove scientific notation
plt.ticklabel_format(style= 'plain')
```
